# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

### Extracting the stats of the matches of all teams which they played in their home ground
- creating a pivot table with index as **"season, hometeam"**.
- aggregate function used is **count** (to count the number of matches won lost and draw)

```python
df_pivot_hometeam_matches = (data.pivot_table(columns=['ftr'], index=['season', 'hometeam'] , aggfunc='count')['awayteam']).reset_index()

df_pivot_hometeam_matches.head()
```

### Extracting the stats of the matches of all teams where they played as away team
- the same process as done to extract above table

```python
df_pivot_awayteam_matches = (data.pivot_table(columns=['ftr'], index=['season', 'awayteam'] , aggfunc='count')['hometeam']).reset_index()

df_pivot_awayteam_matches.head()
```

#### checking whether both the above tables(home team & away team) are in sync
- the team name in df_pivot_awayteam_matches is similar to team name in df_pivot_hometeam_matches

```python
(df_pivot_awayteam_matches['awayteam']==df_pivot_hometeam_matches['hometeam']).value_counts()
```

## Merging both the above tables
<hr/>

- Logic used is that for a particular Team when it is played as Home Team the **Won,Lost and Draw** is shown in the above table as ** H , A and D ** respectively

- Similarly for a particular Team when it is played as Away Team the **Won,Lost and Draw** is shown in the above table as ** A , H and D ** respectively

- Initialized the new data frame **data_new** with **df_pivot_hometeam_matches**

- Appending the new columns to **data_new** from **df_pivot_awayteam_matches** using the logic mentioned as in above 2 poinnts

- Renaming the columns for ease of usage and to bring clarity and visual sense to it

- Outcome of the Matches henceforth is represented as **outcome_HT or outcome_AT** (for eg lost_HT: it is nothing but match lost being home team and won_AT is nothing but Match won as Away Team)

```python
data_new = df_pivot_hometeam_matches

#matches lost when played as away team
data_new['lost'] = df_pivot_awayteam_matches['H'] 

#matches which were draw when played as away team
data_new['draw'] = df_pivot_awayteam_matches['D']

#matches won when played as away team
data_new['won'] = df_pivot_awayteam_matches['A']

data_new.columns = ['Season', 'Team', 'lost_HT', 'draw_HT', 'won_HT', 'lost_AT', 'draw_AT', 'won_AT']
```

```python
data_new.head()
```

### To get the values for the Lost and Won into one column
<hr/>

- The idea is to combine the total **Win,Loss & Draw** column values under one column

```python
data_won_lost = data_new.iloc[:, :2]

data_won_lost['won'] = data_new['won_HT']+data_new['won_AT']

data_won_lost['lost'] = data_new['lost_HT']+data_new['lost_AT']

data_won_lost['draw'] = data_new['draw_HT']+data_new['draw_AT']

data_won_lost['Season'] = data_new['Season'].apply(lambda x : (int)(x[:4]))

data_won_lost = data_won_lost.replace(np.nan, 0)

data_won_lost
```

### Sorting the above table basis most number of Wins in a particular season
<hr/>

- Used the custom sort on columns **Season, Won**
- "Column Season" sorted in the ascending order and
- "Column Won" sorted in the descending order

```python
Season_teamRanking = data_won_lost.sort_values(by=['Season', 'won'], ascending=[True,False], ignore_index=True).set_index(['Season', 'Team'])
Season_teamRanking
```

### TOP 5 TEAMS SEASON WISE
<hr/>

- Extracting the **Season data** from the above resultant table
- Then **Top 5 Teams data** is being extracted from each Season's data
- The names of Top 5 Teams from each Season is pushed into the list (top_5_list)
- From the list the Team with highest frequency of occurnces in top_5_list will be the Top and the next highest frequence will be second and so on

```python
seasons = list(range(1995, 2018))

top_5_list = []

top_5 = pd.DataFrame(index=seasons)
for season in seasons:
    print('\n\nSeaons : ', (str)(season), '-', (str)(season)[2:], sep='')
    obj = Season_teamRanking.loc[season].iloc[:5]
    print(obj)
    top_5_list += list(obj.index.values)

# print(top_5_list)
top_5_list = np.array(top_5_list);

teams, freq = np.unique(top_5_list, return_counts=True)
consistent_teams = pd.DataFrame(data={'teams':teams, 'freq':freq}).sort_values(by='freq', ascending=False).set_index('teams').iloc[:5]

top5_consistent = consistent_teams.index.values
top5_consistent
```

### Season wise performance of individual teams
<hr/>

- To get the table for "Team's performance" **Season wise** for the overall 23 years

```python
Seasonwise_performance = data_won_lost.sort_values(by=['Team', 'Season'], ascending=[True, True]).set_index(['Team','Season'])
Seasonwise_performance
```

### Extracting the **Season wise** data for the Top 5 Teams
<hr/>

- The columns in the below table represents the Top 5 Teams while the rows represent the number of wins in the respective Season

```python
graph_data = pd.DataFrame(index=seasons)
for team in top5_consistent:
    var1 = np.zeros((23, 1))
    var2 = np.array(Seasonwise_performance.loc[team].loc[:, ['won']].values)
    var1[:var2.shape[0], :] = var2[:, :]
    graph_data[team] = var1
    
graph_data.head()
```

### To observer the performance of the Top 5 teams over 23 seasons

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(seasons, graph_data)
ax.set_xticks(seasons)
ax.legend(loc = 'lower center', labels=graph_data.columns, ncol=2, labelspacing=1.2)

plt.show()
```

### Extracting the Top 10 Teams for the overall 23 years Season
<hr/>

- The logic and process is similar to the one explained above for extracting Top 5 Teams

```python
top_10_list = []

for season in seasons:
#     print('\n\nSeaons : ', (str)(season), '-', (str)(season)[2:], sep='')
    obj = Season_teamRanking.loc[season].iloc[:10]
#     print(obj)
    top_10_list += list(obj.index.values)

# print(top_10_list)
top_10_list = np.array(top_10_list);

teams, freq = np.unique(top_10_list, return_counts=True)
consistent_teams = pd.DataFrame(data={'teams':teams, 'freq':freq}).sort_values(by='freq', ascending=False).set_index('teams').iloc[:10]

top10_consistent = consistent_teams.index.values
top10_consistent
```

#### For the Top 10 Teams , extracting the number of Matches Won,Lost and Draw along with that Win ratio (matches won/Total Matches played) is also calculated

```python
data_10_25Seasons = data_won_lost.groupby('Team').agg(np.sum).drop('Season', axis=1).loc[top10_consistent]

data_10_25Seasons['percentage'] = np.round(((data_10_25Seasons['won'])/(data_10_25Seasons['won']+data_10_25Seasons['lost']+data_10_25Seasons['draw']))*100,2)

data_10_25Seasons = data_10_25Seasons.sort_values(by='percentage', ascending=False)

data_10_25Seasons
```

### Bar Chart of Win Ratio for Top 10 Teams

```python
fig, ax = plt.subplots(figsize = (15,6))

ax = sns.barplot(x=data_10_25Seasons.index, y=data_10_25Seasons.percentage, ax=ax)

ax.set_xticklabels(labels=data_10_25Seasons.index, rotation=90, fontsize=14)

ax.set_title(label='Percentage of Matches won by \ntop_10_consistent teams over 25 seasons',
                    fontsize=16, color='brown', loc='right')

for bar in ax.patches:
    ax.annotate(format(bar.get_height(), '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0, 5), 
                   textcoords='offset points') 
    
ax.spines['top'].set_visible(False)
ax.spines["left"].set_visible(False) 
ax.spines["right"].set_visible(False) 
ax.axes.yaxis.set_visible(False) 

plt.show()
```

### Extracting the data to show the converstion percentage of the Top 10 Teams
<hr/>

- **Coverstion Percentage** : Teams where half time score was less than the opponent Team however they won as per Full tiem result
- it is basically to turn around the Matches from Losing to Winning

```python
data_ftr_To_htr = data[(data['hometeam'].isin(top10_consistent)) | (data['awayteam'].isin(top10_consistent))]

data_ftr_To_htr = data_ftr_To_htr[['hometeam','awayteam','htr','ftr']]

data_ftr_To_htr
```

```python
bar2_data = pd.DataFrame(index=top10_consistent, columns=['htrLost_ftrWon'])

for team in top10_consistent:
    cond1 = (data_ftr_To_htr['hometeam']==team) & (data_ftr_To_htr['htr']!='H') & (data_ftr_To_htr['ftr']=='H')
    cond2 = (data_ftr_To_htr['awayteam']==team) & (data_ftr_To_htr['htr']!='A') & (data_ftr_To_htr['ftr']=='A')
    bar2_data.loc[team] = data_ftr_To_htr[cond1 | cond2].shape[0]

bar2_data
```

### Bar Chart for the converstion ratio for Top 10 Teams

```python
fig, htr_ftr_ax = plt.subplots(figsize=(15, 6))

sns.barplot(bar2_data.index, bar2_data.htrLost_ftrWon, ax=htr_ftr_ax)

htr_ftr_ax.set_xticklabels(labels=bar2_data.index, rotation=90, fontsize=14)

htr_ftr_ax.set_title(label='Matches of Top_10_consistent teams over 25 seasons\nwhere they could convert Half Time Loss to Full time win',
                    fontsize=16, color='brown', loc='right')

for bar in htr_ftr_ax.patches:
    htr_ftr_ax.annotate(format(bar.get_height(), '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0, 5), 
                   textcoords='offset points')
    
htr_ftr_ax.spines['top'].set_visible(False)
htr_ftr_ax.spines["left"].set_visible(False) 
htr_ftr_ax.spines["right"].set_visible(False) 
htr_ftr_ax.axes.yaxis.set_visible(False) 

plt.show()
```
